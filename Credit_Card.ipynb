```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv'
```

```python
df = pd.read_csv(url)
df.head()
```

```python
# --- Count missing values ---
print("Missing Values Count:")
df.isnull().sum()
```

```python
# --- Show top 3 correlations ---
print("\nTop 3 Negative Correlations with 'Class':")
top_correlations = df.corr()['Class'].sort_values(ascending=True).head(3)
top_correlations
```

```python
# --- Plot one histogram ---
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Amount'] < 500]['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts (Under $500)')
plt.xlabel('Transaction Amount ($)')
plt.ylabel('Frequency')
plt.show()
```
